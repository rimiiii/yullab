```python
from openai import OpenAI
import os
```

```python
client = OpenAI(api_key=os.environ('OPENAI_API_KEY'))
  
assistant = client.beta.assistants.create(
  name="능력있는 소설가",
  instructions="로맨스 소설의 첫 문장을 입력하면 그 문장 다음으로 나올 글을 한 문단 작성해주세요.",
  # tools=[{"type": "code_interpreter"}],
  model="gpt-4-turbo-preview",
)
```

```python
thread = client.beta.threads.create()
```

```python
message = client.beta.threads.messages.create(
    thread_id=thread.id,
    role='user',
    content='등록금을 벌기 위해서 친한 형의 카페에서 일하며 취미로 사진 찍는 걸 좋아하는 대학생 지완. 어느 날 지완은 카메라를 만지던 중 렌즈 안에 비친 수인을 보게 되고 몰래 사진을 찍으려다 경희에게 들키게 된다.'
)
```

```python
run = client.beta.threads.runs.create(
    thread_id=thread.id,
    assistant_id=assistant.id,
    instructions="다음 이야기가 이어질 수 있게 스토리를 풀다가 중간에 끊어주세요. 릴레이로 작성할 글을 위해 주인공이 행동할 수 있는 선택지를 2개 주세요."
)
```

```python
run = client.beta.threads.runs.retrieve(
  thread_id=thread.id,
  run_id=run.id
)
```

```python
import time

def poll_run(run, thread):
    while run.status != 'completed':
        run = client.beta.threads.runs.retrieve(
            thread_id=thread.id,
            run_id=run.id,
        )
        time.sleep(0.5)
    return run

run = poll_run(run, thread)
```

```python
messages = client.beta.threads.messages.list(
  thread_id=thread.id
)
```

```python
for m in messages:
    print(f'{m.role}: {m.content[0].text.value}')
```

```python
run = client.beta.threads.runs.create(
    thread_id=thread.id,
    assistant_id=assistant.id,
    instructions="제시한 선택지에서 `2. 경희의 불안과 부담감을 이해하고 전시회 참여를 포기하여, 두 사람의 관계를 우선시하는 길.`를 선택했습니다. 이 뒤에 올 행동을 한 문단으로 작성해주세요. 다음 이야기가 이어질 수 있게 스토리를 풀다가 중간에 끊어주세요. 릴레이로 작성할 글을 위해 주인공이 행동할 수 있는 선택지를 2개 주세요."
)

run = client.beta.threads.runs.retrieve(
  thread_id=thread.id,
  run_id=run.id
)

run = poll_run(run, thread)

messages = client.beta.threads.messages.list(
  thread_id=thread.id
)

for m in messages:
    print(f'{m.role}: {m.content[0].text.value}')
```
